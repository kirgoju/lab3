**Задание 1**

Напишите код, который строит график функции y = x^2 на интервале [−10, 10].

**7. Трехмерный график поверхности.**

Используя mpl_toolkits.mplot3d , постройте 3D-график функции $z = \sin(\sqrt{x^2 + y^2})$
на сетке значений x, y в диапазоне [−5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf, ax=ax)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'3D-график поверхности')
plt.show()
```

**Индивидуально задание 3. Вариант 20.**

$$f(x) = x^2 \cdot e^{-x}$$
на интервале [0, 3].

1. Построить график подинтегральной функции.
2. Вычислить площадь под кривой на заданном отрезке как значение
определенного интеграла.
3. Закрасить область под графиком, чтобы визуализировать интеграл.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.integrate import quad
x = np.linspace(0, 3, 50)
y = (x**2)*(math.e**-x)
plt.fill_between(x, y, alpha=0.5)
def f(x):
    return x**2 * np.exp(-x)
p = quad(f, 0, 3)
print(f"Значение интеграла (площадь под кривой): {p}")
plt.plot(x,y)
plt.title('f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

**Индивидуальное задание 4. Вариант 20**
$$f(x, y) = e^{-x^{2} - y^{2}} \cos(5 (x^{2} + y^{2})$$
на интервале x, y ∈ [-3, 3].

- Используйте plot_surface() для плотных поверхностей.
- Используйте plot_wireframe() для каркасных моделей.
- Настроите цветовую карту через cmap .
- Добавьте подписи осей и заголовки.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
def f(x, y):
    return np.exp(-x**2 - y**2) * np.cos(5 * (x**2 + y**2))
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.8)
fig.colorbar(surf, ax=ax1, label='Значение f(x, y)')
ax1.set_title('Поверхность')
ax1.set_xlabel('x', fontsize=10)
ax1.set_ylabel('y', fontsize=10)
ax1.set_zlabel('f(x, y)', fontsize=10)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z, color='darkblue', linewidth=0.5)
ax2.set_title('Каркас')
ax2.set_xlabel('x' )
ax2.set_ylabel('y' )
ax2.set_zlabel('f(x, y)' )
plt.tight_layout()
plt.show()
```
